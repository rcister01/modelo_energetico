## Armamos lista de features temporales

## Armamos dataframe nuevo con feautres atemporales y temporales reducidos

```python
X_train_f = X_train.iloc[:, 0:20]
X_train_f.head()
```

```python
X_train_s = pd.DataFrame()
for column in columns_s:
    morning = X_train[f'{column}_0']
    afternoon = 0
    night = 0
    
    time_of_day = 'm'
    
    for hour in range(1,672):
        if hour%8 == 0:
            if time_of_day == 'm':
                time_of_day = 'a'
            elif time_of_day == 'a':
                time_of_day = 'n'
            else:
                time_of_day = 'm'
            
        if time_of_day == 'm':
            morning = morning + X_train[f'{column}_{hour}']
        elif time_of_day == 'a':
            afternoon = afternoon + X_train[f'{column}_{hour}']
        else:
            night = night + X_train[f'{column}_{hour}']
    
    X_train_s[f'{column}_morning'] = morning/672
    X_train_s[f'{column}_afternoon'] = afternoon/672
    X_train_s[f'{column}_night'] = night/672
                                    
```

```python
X_train_s.head()
```

```python
X_train_red = pd.concat([X_train_f, X_train_s], axis=1)
X_train_red.head()
```
